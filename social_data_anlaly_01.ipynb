```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            firstNum = nums[i]
            for j in range(1, len(nums)):
                secondNum = nums[j]
                if firstNum + secondNum == target:
                    return [i,j]
        return [-1, -1]
```

## leetcode 문제_02

```python
# 이 문제는 Add Two Numbers 문제로, 두 연결 리스트에 저장된 숫자를 더하는 문제입니다.
# 숫자가 역순으로 저장되어 있기 때문에 자리 올림(carry) 을 처리하면서 노드를 하나씩 더하면 됩니다.
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)  # 결과 리스트의 시작점
        current = dummy
        carry = 0

        while l1 or l2 or carry:
            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0

            total = v1 + v2 + carry
            carry = total // 10

            current.next = ListNode(total % 10)
            current = current.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def build_list(arr):
    dummy = ListNode(0)
    cur = dummy

    for v in arr:
        cur.next = ListNode(v)
        cur = cur.next

    return dummy.next


class Solution:
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)
        cur = dummy
        carry = 0

        while l1 or l2 or carry:

            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0

            total = v1 + v2 + carry
            carry = total // 10

            cur.next = ListNode(total % 10)
            cur = cur.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next


l1 = build_list([9,9,9,9,9,9,9])
l2 = build_list([9,9,9,9])

sol = Solution()
result = sol.addTwoNumbers(l1, l2)

while result:
    print(result.val, end=" ")
    result = result.next
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import MultiComparison

# 데이터 불러오기
df = pd.read_csv("직무만족도.txt", sep="\t")

# -----------------------------
# 문제1
# -----------------------------

# (1) 성별 빈도 및 퍼센트
gender_freq = df['gender'].value_counts().sort_index()
gender_pct = df['gender'].value_counts(normalize=True).sort_index() * 100

print("성별 빈도\n", gender_freq)
print("성별 퍼센트\n", gender_pct)

# (2) 연령 범주화
def age_group(age):
    if age <= 39:
        return "30대 이하"
    elif age <= 49:
        return "40대"
    else:
        return "50대 이상"

df['age_group'] = df['age'].apply(age_group)

age_freq = df['age_group'].value_counts()
age_pct = df['age_group'].value_counts(normalize=True) * 100

print("\n연령대 빈도\n", age_freq)
print("연령대 퍼센트\n", age_pct)


# -----------------------------
# 문제2 (성별에 따른 Sat 차이: t-test)
# -----------------------------

# (1) 가설
# H0: μ1 = μ2
# H1: μ1 ≠ μ2

# (2) 평균
print("\n성별 평균")
print(df.groupby('gender')['Sat'].mean())

# (3) t-test
g1 = df[df['gender']==1]['Sat']
g2 = df[df['gender']==2]['Sat']

t_stat, p_val = stats.ttest_ind(g1, g2, equal_var=True)

print("\nT-test 결과")
print("t값:", t_stat)
print("p값:", p_val)


# -----------------------------
# 문제3
# -----------------------------

# S5 역코딩 (최대값 9 기준이라고 가정)
df['S5_rev'] = 10 - df['S5']

# 직무내적(Y1): S1,S2,S3
df['Y1'] = df[['S1','S2','S3']].mean(axis=1)

# 직무외적(Y2): S4,S5_rev
df['Y2'] = df[['S4','S5_rev']].mean(axis=1)

# (1) 평균
print("\nY1 평균:", df['Y1'].mean())
print("Y2 평균:", df['Y2'].mean())

# (2) 상관계수
corr = df['Y1'].corr(df['Y2'])
print("\n상관계수:", corr)


# -----------------------------
# 문제4 (연령대별 Sat 차이: ANOVA)
# -----------------------------

# (1) 가설
# H0: μ1 = μ2 = μ3
# H1: 적어도 하나 다름

# (2) 기술통계
print("\n연령대별 Sat 기술통계")
print(df.groupby('age_group')['Sat'].describe())

# (3) ANOVA
model = ols('Sat ~ C(age_group)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nANOVA 결과")
print(anova_table)

# (5) Duncan 대신 Tukey (Python 기본)
mc = MultiComparison(df['Sat'], df['age_group'])
tukey = mc.tukeyhsd()

print("\n사후분석 (Tukey)")
print(tukey)


# -----------------------------
# 문제5 (교차분석 + 카이제곱)
# -----------------------------

# 교차표 (행 퍼센트)
ct = pd.crosstab(df['age_group'], df['period'], normalize='index') * 100
print("\n교차표 (행 퍼센트)")
print(ct)

# 카이제곱 검정
ct_raw = pd.crosstab(df['age_group'], df['period'])
chi2, p, dof, exp = stats.chi2_contingency(ct_raw)

print("\n카이제곱 검정")
print("chi2:", chi2)
print("p값:", p)


# -----------------------------
# 문제6 (중회귀분석)
# -----------------------------

X = df[['Y1','Y2','age']]
X = sm.add_constant(X)
y = df['Sat']

model = sm.OLS(y, X).fit()

print("\n회귀분석 결과")
print(model.summary())

# 회귀식 출력
coef = model.params
print("\n회귀식:")
print(f"Sat = {coef[0]:.3f} + {coef[1]:.3f}*Y1 + {coef[2]:.3f}*Y2 + {coef[3]:.3f}*age")
```

## 1_1 지민 직무만족도를 pytorch로 구현

```python
# =========================
# 1. GPU 설정
# =========================
import torch
import pandas as pd
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("사용 장치:", device)
```

```python
# =========================
# 2. 데이터 불러오기
# =========================
# Colab에 파일 업로드 후 실행
from google.colab import files
# uploaded = files.upload()

df = pd.read_csv("/content/직무만족도.txt", sep="\t")
df.head()
```

```python
# tensor 변환 + GPU 이동
data = torch.tensor(df.values, dtype=torch.float32).to(device)

cols = {name: i for i, name in enumerate(df.columns)}
```

```python
# =========================
# 문제1
# =========================
gender = data[:, cols['gender']]
total = len(gender)

unique, counts = torch.unique(gender, return_counts=True)
print("\n[문제1-성별]")
```

```python
for u, c in zip(unique, counts):
    print(f"{int(u.item())}: {int(c.item())} ({c.item()/total*100:.2f}%)")
```

```python
# (2) 연령 범주화
age = data[:, cols['age']]

age_group = []
for a in age:
    if a <= 39:
        age_group.append(0)  # 30대 이하
    elif a <= 49:
        age_group.append(1)  # 40대
    else:
        age_group.append(2)  # 50대 이상

age_group = torch.tensor(age_group)

print("\n연령대 빈도")
for g in [0, 1, 2]:
    count = (age_group==g).sum()
    print(f"{g}: {int(count)} ({count/total*100:.2f}%)")
```

```python
# -----------------------------
# 문제2 (t-test 직접 구현)
# -----------------------------
Sat = data[:, cols['Sat']]

g1 = Sat[gender==1]
g2 = Sat[gender==2]

mean1, mean2 = g1.mean(), g2.mean()
var1, var2 = g1.var(unbiased=True), g2.var(unbiased=True)

n1, n2 = len(g1), len(g2)

print(mean1, mean2)
print('\n', var1, var2)
print('\n', n1, n2)
```

```python
# pooled variance
sp = ((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2)

# t-stat
t_stat = (mean1 - mean2) / torch.sqrt(sp*(1/n1 + 1/n2))
print("\n문제2 t값:", t_stat.item())
```

```python
# -----------------------------
# 문제3
# -----------------------------
S1 = data[:, cols['S1']]
S2 = data[:, cols['S2']]
S3 = data[:, cols['S3']]
S4 = data[:, cols['S4']]
S5 = data[:, cols['S5']]

# 역코딩 (7점 기준)
S5_rev = 8 - S5

Y1 = torch.stack([S1, S2, S3]).mean(dim=0)
Y2 = torch.stack([S4, S5_rev]).mean(dim=0)

print("Y1 평균:", Y1.mean().item())
print("Y2 평균:", Y2.mean().item())
```

```python
# 상관계수 (피어슨)
def pearson(x, y):
    xm = x - x.mean()
    ym = y - y.mean()
    return (xm*ym).sum() / torch.sqrt((xm**2).sum()*(ym**2).sum())
```

```python
corr = pearson(Y1, Y2)
print("상관계수:", corr.item())
```

```python
# -----------------------------
# 문제4 (ANOVA 직접 구현)
# -----------------------------
groups = [Sat[age_group==i] for i in [0,1,2]]

overall_mean = Sat.mean()

# Between
SSB = sum([len(g)*(g.mean()-overall_mean)**2 for g in groups])

# Within
SSW = sum([((g - g.mean())**2).sum() for g in groups])

dfb = 3-1
dfw = len(Sat)-3

MSB = SSB / dfb
MSW = SSW / dfw

F = MSB / MSW
print("\nANOVA F값:", F.item())
```

```python
# -----------------------------
# 문제5 (카이제곱)
# -----------------------------
# GPU → CPU 이동 (핵심!!)
period = data[:, cols['period']]

period_cpu = period.cpu()
age_group_cpu = age_group.cpu()

# 교차표 만들기
ct = {}
for a in [0,1,2]:
    for p in torch.unique(period_cpu):
        ct[(a, int(p.item()))] = int(((age_group_cpu == a) & (period_cpu == p)).sum())

print("교차표")
print(ct)
```

```python
# (간단 chi-square 계산)
# 기대값 계산용 행/열 합

# GPU → CPU 이동
age_group_cpu = age_group.cpu()
period_cpu = period.cpu()

row_sum = [(age_group_cpu==i).sum().item() for i in [0,1,2]]
col_values = torch.unique(period_cpu)
col_sum = [(period_cpu==c).sum().item() for c in col_values]

N = len(period_cpu)

chi2 = 0
for i,a in enumerate([0,1,2]):
    for j,c in enumerate(col_values):
        O = ((age_group_cpu==a) & (period_cpu==c)).sum().item()
        E = row_sum[i]*col_sum[j]/N
        chi2 += (O-E)**2/E

print("카이제곱 값:", chi2)
```

```python
# -----------------------------
# 문제6 (회귀분석: PyTorch)
# -----------------------------

X = torch.stack([Y1, Y2, age], dim=1)
X = torch.cat([torch.ones(len(X), 1).to(device), X], dim=1)
y = Sat.view(-1, 1)

beta = torch.inverse(X.T @ X) @ X.T @ y

print("[문제6]")
print("회귀계수:", beta)
```

```python
# R^2
y_hat = X @ beta
SS_tot = ((y - y.mean())**2).sum()
SS_res = ((y - y_hat)**2).sum()
R2 = 1 - SS_res/SS_tot

print("R^2:", R2.item())
```

```python
print("\n회귀식:")
print(f"Sat = {beta[0].item():.3f} + {beta[1].item():.3f}*Y1 + {beta[2].item():.3f}*Y2 + {beta[3].item():.3f}*age")
```
